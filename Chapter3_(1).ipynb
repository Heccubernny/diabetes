```python
# Import the library for ANOVA test
from scipy.stats import f_oneway

# Perform the ANOVA test
category_groups = [cdc_dataset[num_var][cdc_dataset['Diabetes_012'] == category] for category in cdc_dataset["Diabetes_012"].unique()]
F, p = f_oneway(*category_groups)

# Define your significance level (e.g., 0.05)
significance_level = 0.05
category_name = ['non-diabetic','prediabetic','diabetic']
# Interpret the results for each category
for i, category in enumerate(cdc_dataset["Diabetes_012"].unique()):
    if p[i] < significance_level:
        print(f"The means of {category_name[category]} people are significantly different from the others.")
    else:
        print(f"The means of {category_name[category]} people are not significantly different from the others.")
```

An analysis of variance was performed to understand the association between the numerical variables and Diabetes status. The ANOVA test, which is the hypothesis test used in this code, determines whether there are statistically significant differences in the means of the numerical variable across the many categories of the target variable. The alternative hypothesis (H1) states that there are substantial differences between at least one group and the others, contrary to the null hypothesis (H0), which states that there are no significant differences. The significance of the results is evaluated using the p-value (p).

```python
from scipy.stats import chi2_contingency
import numpy as np

# Create a Series containing all the categorical variables
cat_var_data = cdc_dataset.drop('Diabetes_012', axis=1).select_dtypes(include=['category'])

# List of categorical variable names
cat_var_names = cat_var_data.columns

# DataFrame to store the results
results = []

for cat_var_name in cat_var_names:
    # Create a contingency table for each categorical variable
    contingency_table = pd.crosstab(cat_var_data[cat_var_name], cdc_dataset['Diabetes_012'])

    # Calculate Chi-square statistics
    chi2, _, _, _ = chi2_contingency(contingency_table)

    # Calculate Cramr's V
    n = contingency_table.sum().sum()
    v = np.sqrt((chi2 / (n * min(contingency_table.shape[0] - 1, contingency_table.shape[1] - 1))))

    # Store the results in a DataFrame
    results.append({
        'Categorical Variable': cat_var_name,
        'Cramr\'s V': v
    })

# Display the results
result_df = pd.DataFrame(results)
print(result_df)
```

```python
# Load data
cdc_dataset.head()
```

```python
# Import relevant library
from sklearn.ensemble import IsolationForest
# Instantiate IsolationForest
model = IsolationForest(random_state=42)
# Fit on data
model.fit(cdc_dataset)
# Get the anomalies and add as a column to the dataset
cdc_dataset['anomaly']= model.predict(cdc_dataset)
# Anomalies are labeled as -1, and inliers (normal data points) are labeled as 1
cdc_dataset.drop(cdc_dataset[cdc_dataset['anomaly']==-1].index,inplace = True)
# Drop anomaly Column
cdc_dataset.drop(columns=['anomaly'], inplace=True)
```

The IsolationForest model is instantiated with a random state set to 42.

The model is fitted to the "cdc_dataset" data.

Anomalies are predicted within the dataset, and the model's anomaly predictions are added as a new column labeled 'anomaly.'

Rows in the dataset where anomalies are labeled as -1 are removed, effectively filtering out the data points considered anomalies by the IsolationForest model.

Finally, the 'anomaly' column is dropped from the dataset, leaving the dataset cleaned of the additional column used for anomaly labeling. The resulting dataset only contains inliers, representing normal data points as determined by the IsolationForest model.

```python
# Seperate inputs from target
features = cdc_dataset.drop('Diabetes_012', axis=1)
target = cdc_dataset['Diabetes_012']
```

The data was split into features(independent variables) and target (dependent variable)

A pipeline is created that takes in features, it OneHotEncodes the categorical variables and standardizes the numerical variables. Then we apply it on the features.



```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer

cat_names = features.select_dtypes(include=['category']).columns
num_names = features.select_dtypes(exclude=['category']).columns

# Create a custom transformer for LabelEncoding multiple columns
class MultiLabelEncoder:
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X_encoded = X.copy()
        for col in range(X.shape[1]):
            le = LabelEncoder()
            X_encoded[col] = le.fit_transform(X[col])
        return X_encoded

# Create transformers for each feature type
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', MultiLabelEncoder())
])

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

# Combine the transformers into a preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, num_names),
        ('cat', categorical_transformer, cat_names)
    ])

# Create the main pipeline for the entire data preprocessing
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Fit and transform the data using the pipeline
features_preprocessed = pipeline.fit_transform(features)
```

```python
from sklearn.model_selection import train_test_split
# Split data into train and test set
X_train, X_test, y_train, y_test = train_test_split(features_preprocessed, target, test_size=0.2,
                                                    stratify=target, random_state=42)
```

### **3.4 Resampling**  
Resampling is a technique used to modify the distribution of a dataset by changing the number of instances or samples in specific categories or classes. There are two primary resampling methods:

1. **Oversampling:** This involves increasing the number of instances in the minority class or category to balance it with the majority class. This is typically done by duplicating existing instances or generating synthetic samples to create a more balanced dataset.

2. **Undersampling:** Undersampling aims to reduce the number of instances in the majority class to match the number of instances in the minority class. This can be achieved by randomly selecting a subset of instances from the majority class or by using more sophisticated techniques.

Resampling is commonly used when dealing with imbalanced datasets, where one class significantly outweighs the other(s). By adjusting the class distribution, resampling can help machine learning models better learn and make predictions, particularly for the minority class, leading to improved model performance in situations where class imbalance is a challenge.


The algorithm that will be used for resampling the data set is called SMOTEENN.

SMOTEENN stands for Synthetic Minority Over-sampling Technique with Edited Nearest Neighbors. It is a combination of two resampling techniques used in machine learning to address class imbalance in datasets, particularly in binary classification problems:

1. **SMOTE (Synthetic Minority Over-sampling Technique):** SMOTE generates synthetic samples for the minority class by interpolating between existing instances. This helps balance the class distribution by creating additional instances of the minority class.

2. **ENN (Edited Nearest Neighbors):** ENN is used to remove noisy and potentially misclassified data points. It does so by examining the class labels of a data point's nearest neighbors and removing it if its label doesn't match the majority of its neighbors.

By combining SMOTE and ENN, SMOTEENN first oversamples the minority class using SMOTE and then applies ENN to remove noisy or irrelevant samples from both the minority and majority classes. This results in a more balanced dataset while reducing potential noise in the data, making it beneficial for improving the performance of machine learning models, especially in cases of severe class imbalance.

```python
# import SMOTE
from imblearn.combine import SMOTEENN
from imblearn.over_sampling import SMOTE
# Convert the categorical target class to numerical labels

# Initialize the SMOTE object with custom sampling strategy
smote = SMOTEENN(random_state=42)

# Apply SMOTE to create synthetic samples
X_resampled, y_resampled = smote.fit_resample(features_preprocessed, target)
# Split into train and test set
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, random_state=42, stratify = y_resampled)
# Check the new class distribution
unique, counts = np.unique(y_resampled, return_counts=True)
print("Class distribution after SMOTE:")
print(dict(zip(unique, counts)))
```

## **3.5 Hyperparameter Tuning**
Hyperparameter tuning, also known as hyperparameter optimization, is the process of systematically searching for the best combination of hyperparameters for a machine learning model. Hyperparameters are settings or configurations that are not learned from the training data but are set prior to the model's training. They influence the model's behavior and its ability to learn from the data.

The goal of hyperparameter tuning is to find the hyperparameters that result in the best model performance on a validation dataset or through cross-validation. This can include parameters such as learning rates, regularization strengths, the number of hidden layers and units in a neural network, and many others, depending on the specific algorithm or model being used.

Hyperparameter tuning is a critical step in building effective machine learning models because it helps optimize a model's performance, improve its generalization to new data, and achieve better results for real-world applications. Various techniques and tools, such as grid search, random search, and Bayesian optimization, are used to explore and fine-tune hyperparameter values.

We plan to conduct hyperparameter tuning using Randomized Search Cross-Validation (RandomSearchCV) to optimize the performance of three different machine learning models: the RandomForestClassifier, XGBoost Classifier (XGBClassifier), and K-Nearest Neighbors Classifier (KNeighborClassifier). The primary objective is to identify models that demonstrate strong generalization capabilities when applied to new data.

For XGBoost Classifier (XGBClassifier):

- 'n_estimators': The number of boosting rounds or decision trees in the ensemble.
- 'max_depth': The maximum depth of each decision tree.
- 'learning_rate': The step size at which the model adapts based on errors.
- 'subsample': The fraction of samples used for training each tree.
- 'colsample_bytree': The fraction of features (columns) used for training each tree.
- 'min_child_weight': Minimum sum of instance weight (hessian) needed in a child.
- 'gamma': Minimum loss reduction required to make a further partition on a leaf node.
- 'alpha': L1 regularization term on weights.
- 'lambda': L2 regularization term on weights.

For RandomForest Classifier:

- 'n_estimators': The number of decision trees in the forest.
- 'max_depth': The maximum depth of each decision tree in the forest.
- 'min_samples_split': The minimum number of samples required to split an internal node.
- 'min_samples_leaf': The minimum number of samples required to be at a leaf node.
- 'max_features': The number of features to consider when looking for the best split.

For K-Nearest Neighbors Classifier (KNeighborClassifier):

- 'n_neighbors': The number of neighbors to consider for classification.
